### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda1 = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

# Mostrar los resultados
print(f"Ingreso total de la Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Ingreso total de la Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Ingreso total de la Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Ingreso total de la Tienda 4: ${ingreso_tienda4:,.2f}")
```

# 2. Ventas por categoría

```python
# Contar productos vendidos por categoría en cada tienda
ventas_por_categoria1 = tienda["Categoría del Producto"].value_counts()
ventas_por_categoria2 = tienda2["Categoría del Producto"].value_counts()
ventas_por_categoria3 = tienda3["Categoría del Producto"].value_counts()
ventas_por_categoria4 = tienda4["Categoría del Producto"].value_counts()

# Mostrar los resultados
print("Tienda 1 - Ventas por categoría:")
print(ventas_por_categoria1)
print("\nTienda 2 - Ventas por categoría:")
print(ventas_por_categoria2)
print("\nTienda 3 - Ventas por categoría:")
print(ventas_por_categoria3)
print("\nTienda 4 - Ventas por categoría:")
print(ventas_por_categoria4)
```

# 3. Calificación promedio de la tienda


```python
# Calcular la calificación promedio para cada tienda
calificacion_promedio1 = tienda["Calificación"].mean()
calificacion_promedio2 = tienda2["Calificación"].mean()
calificacion_promedio3 = tienda3["Calificación"].mean()
calificacion_promedio4 = tienda4["Calificación"].mean()

# Mostrar resultados
print(f"Calificación promedio de la Tienda 1: {calificacion_promedio1:.2f}")
print(f"Calificación promedio de la Tienda 2: {calificacion_promedio2:.2f}")
print(f"Calificación promedio de la Tienda 3: {calificacion_promedio3:.2f}")
print(f"Calificación promedio de la Tienda 4: {calificacion_promedio4:.2f}")
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt

# Función para identificar y graficar los productos más y menos vendidos
def productos_mas_menos_vendidos(df, nombre_tienda):
    # Contar las ventas de cada producto
    ventas_por_producto = df["Producto"].value_counts()

    # Obtener los productos más vendidos (primeros 5)
    mas_vendidos = ventas_por_producto.head(5)

    # Obtener los productos menos vendidos (últimos 5)
    menos_vendidos = ventas_por_producto.tail(5)

    # Imprimir resultados
    print(f"\n{nombre_tienda} - Productos más vendidos:")
    print(mas_vendidos)
    print(f"\n{nombre_tienda} - Productos menos vendidos:")
    print(menos_vendidos)

    # Graficar los resultados
    fig, axs = plt.subplots(1, 2, figsize=(14, 5))

    # Gráfico de productos más vendidos
    mas_vendidos.plot(kind='bar', ax=axs[0], color='green')
    axs[0].set_title(f'{nombre_tienda} - Más vendidos')
    axs[0].set_ylabel("Cantidad vendida")
    axs[0].set_xticklabels(mas_vendidos.index, rotation=45, ha='right')

    # Gráfico de productos menos vendidos
    menos_vendidos.plot(kind='bar', ax=axs[1], color='red')
    axs[1].set_title(f'{nombre_tienda} - Menos vendidos')
    axs[1].set_ylabel("Cantidad vendida")
    axs[1].set_xticklabels(menos_vendidos.index, rotation=45, ha='right')

    plt.tight_layout()
    plt.show()

# Aplicar a cada tienda
productos_mas_menos_vendidos(tienda, "Tienda 1")
productos_mas_menos_vendidos(tienda2, "Tienda 2")
productos_mas_menos_vendidos(tienda3, "Tienda 3")
productos_mas_menos_vendidos(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
# Calcular el costo de envío promedio para cada tienda
costo_envio_promedio1 = tienda["Costo de envío"].mean()
costo_envio_promedio2 = tienda2["Costo de envío"].mean()
costo_envio_promedio3 = tienda3["Costo de envío"].mean()
costo_envio_promedio4 = tienda4["Costo de envío"].mean()

# Mostrar los resultados
print(f"Costo de envío promedio de la Tienda 1: {costo_envio_promedio1:.2f}")
print(f"Costo de envío promedio de la Tienda 2: {costo_envio_promedio2:.2f}")
print(f"Costo de envío promedio de la Tienda 3: {costo_envio_promedio3:.2f}")
print(f"Costo de envío promedio de la Tienda 4: {costo_envio_promedio4:.2f}")
```

```python
import matplotlib.pyplot as plt

# 1. Gráfico de barras de ingresos totales por tienda
ingresos_totales = [
    tienda["Precio"].sum(),
    tienda2["Precio"].sum(),
    tienda3["Precio"].sum(),
    tienda4["Precio"].sum()
]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(10, 6))
plt.bar(tiendas, ingresos_totales, color='skyblue')
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingresos ($)')
plt.show()

# 2. Gráfico de barras de la distribución de categorías de productos
# Para Tienda 1
ventas_por_categoria1 = tienda["Categoría del Producto"].value_counts()

plt.figure(figsize=(10, 6))
ventas_por_categoria1.plot(kind='bar', color='lightcoral')
plt.title('Distribución de Categorías de Productos - Tienda 1')
plt.xlabel('Categoría')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45, ha='right')
plt.show()

# 3. Gráfico de líneas de las calificaciones promedio por tienda
calificaciones_promedio = [
    tienda["Calificación"].mean(),
    tienda2["Calificación"].mean(),
    tienda3["Calificación"].mean(),
    tienda4["Calificación"].mean()
]

plt.figure(figsize=(10, 6))
plt.plot(tiendas, calificaciones_promedio, marker='o', color='green', linestyle='-', linewidth=2, markersize=8)
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.ylim(0, 5)  # Asumimos que las calificaciones son de 1 a 5
plt.grid(True)
plt.show()
```

```python
# Crear un dataframe unificado con todas las tiendas
df_total = pd.concat([
    tienda.assign(Tienda="Tienda 1"),
    tienda2.assign(Tienda="Tienda 2"),
    tienda3.assign(Tienda="Tienda 3"),
    tienda4.assign(Tienda="Tienda 4")
])

# Ahora sí el gráfico geográfico
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

for tienda_nombre in df_total["Tienda"].unique():
    datos = df_total[df_total["Tienda"] == tienda_nombre]
    plt.scatter(datos["lon"], datos["lat"], label=tienda_nombre, alpha=0.6, s=20)

plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.title("Distribución geográfica de ventas por tienda")
plt.legend()
plt.grid(True)
plt.show()
```
